Memisahkan data parameter dengan variabel hasil

```python
x = df.iloc[:,0:8]
y = df.iloc[:,8]

y
```

Bagi menjadi dua data yaitu data training, dan data testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.1)
y_train
```

Klasifikasi data dari data train yang di buat tadi

```python
from sklearn.svm import SVC

clf = SVC()
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)

y_pred
```

Menampilkan akurasi data dari yang di buat tadi

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
